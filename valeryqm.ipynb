```
import pandas as pd
dataset = pd.read_csv("/content/https://drive.google.com/drive/)
print(dataset.head())

x = dataset['Father'].values.reshape(-1,1)
y = dataset['Son'].values
```

Ahora ajustemos la regresi√≥n lineal a nuestro modelo

```
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

```
from google.colab import drive
drive.mount('/content/drive')
```

Podemos graficar nuestros datos y compararlos con las predicciones de nuestro modelo.

```
import matplotlib.pyplot as plt

plt.scatter(x, y, color = 'b', alpha = 0.3)
plt.plot(x, lin_reg.predict(x), color = 'red', linewidth=2)
plt.title('Estatura Padre vs Estatura Hijo')
plt.xlabel('Estatura Padre')
plt.ylabel('Estatura Hijo')
plt.show()
```

Vamos a predecir !

```
lin_reg.predict([[165]])
```
